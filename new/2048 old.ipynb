```python
def draw_grid():
    bg = (205, 193, 180)
    grid = (187, 173, 160)
    screen.fill(bg)
    pygame.draw.rect(screen, (187, 173, 160),  (0, 400, 400, 40))
    for x in range(1,4):
        pygame.draw.line(screen, grid, (0, x *100),
                         (screen_width, x * 100), line_width)
        pygame.draw.line(screen, grid, (x * 100, 0),
                         (x * 100, screen_height-20), line_width)
```

```python
field = []

def draw_full():
    full_text = 'The board is full!'
    full_img = font.render(full_text, True, blue)
    pygame.draw.rect(screen, green,  (screen_width // 2 - 100, screen_height // 2 - 60, 200, 50))
    screen.blit(full_img, (screen_width // 2 - 100, screen_height // 2 - 50))
    
    again_text = 'Play again?'
    again_img = font.render(again_text, True, blue)
    pygame.draw.rect(screen, green, again_rect)
    screen.blit(again_img, (screen_width // 2 - 80, screen_height // 2 + 10))

def full_field():
    sum = 1
    for i in field:
        for k in i:
            sum *= k
    if k != 0:
        draw_full()
        if event.type == pygame.MOUSEBUTTONDOWN and clicked == False:
                clicked = True
        if event.type == pygame.MOUSEBUTTONUP and clicked == True:
            clicked = False
            pos = pygame.mouse.get_pos()
            if again_rect.collidepoint(pos):
                line_width = 6
                empty_field()
                clicked = False
                game_over = False
         




def upgrade(value):
    pass

def empty_field():
    for x in range(4):
        row = [0]*4
        field.append(row)

    starter = []
    start_works = True
    while start_works:
        for i in range(4):
            starter.append(random.randint(0, 3))
        point_1 = [starter[0], starter[1]]
        point_2 = [starter[2], starter[3]]
        if point_1 != point_2:
            start_works = False
            point_1.append(random.randint(1,2)*2)
            point_2.append(random.randint(1,2)*2)
            print(point_1, point_2)

    field[point_1[0]][point_1[1]] = point_1[2]
    field[point_2[0]][point_2[1]] = point_2[2]        

    print(field)

empty_field()
```

```python
import math

def draw_markers():
    x_pos = 0
    for x in field:
        y_pos = 0
        for y in x:
            if y == 0:
                pass
            else:
                # color == 255/color
                color = colors[y]
                tile_img = font.render(str(y), True, (0,0,0))
                tile_img.set_alpha(180)
                pygame.draw.rect(screen, (color),  (x_pos*100, y_pos*100, 100, 100))
                screen.blit(tile_img, (x_pos*100 + 10, y_pos*100 + 10))
            y_pos += 1
        x_pos += 1
        
rand_gen = False


def place_new():
    global rand_gen
    while rand_gen:
        x = random.randint(0,3)
        y = random.randint(0,3)
        if field[x][y] == 0:
            n = random.randint(0,9)
            if n == 0:
                n = 4
            else: n = 2
            field[x][y] = n
            print('new ', n, ' on ', x, 'x', y)
            rand_gen = False
            
def update_score(new_sum):
    global score
    score += new_sum
    
```

```python
def limit_check(pos, dir):
    if dir == 1:
        if pos < 3:
            return True
    elif dir == -1:
        if pos > 0:
            return True
    else: return False

def move_horz(dir):
    global rand_gen
    if dir == -1:
        start_pos = 3
    else: start_pos = 0
    x_pos = start_pos
    for column in counter_reverse:
        y_pos = start_pos
        
        for row in counter_reverse:
            print(column, row, field[column][row])
            if field[x_pos][y_pos] != 0:
                if limit_check(x_pos, dir):
                    print(x_pos, dir)
                    if field[x_pos+dir][y_pos] == 0:
                        field[x_pos+dir][y_pos] = field[x_pos][y_pos]
                        field[x_pos][y_pos] = 0
                        rand_gen = True
                    elif field[x_pos+dir][y_pos] == field[x_pos][y_pos] and field[x_pos][y_pos] !=0:
                        field[x_pos+dir][y_pos] = 2*field[x_pos][y_pos]
                        update_score(2*field[x_pos][y_pos])
                        field[x_pos][y_pos] = 0
                        rand_gen = True
            y_pos += dir
        x_pos += dir
    for i in counter: 
        x_pos = start_pos   
        for column in counter_reverse:
            y_pos = start_pos
            for row in counter_reverse:
                if field[x_pos][y_pos] != 0:
                    if limit_check(x_pos, dir):
                        if field[x_pos+dir][y_pos] == 0:
                            field[x_pos+dir][y_pos] = field[x_pos][y_pos]
                            field[x_pos][y_pos] = 0
                            rand_gen = True
                y_pos += dir
            x_pos += dir
    place_new()
    
def move_vert(dir):
    global rand_gen

    if dir == -1:
        start_pos = 3
    else: start_pos = 0
    x_pos = start_pos
    for column in counter:
        y_pos = start_pos
        for row in counter:
            print('---',[x_pos], [y_pos])
            if field[x_pos][y_pos] != 0:
                if limit_check(y_pos, dir):
                    if field[x_pos][y_pos+dir] == 0:
                        field[x_pos][y_pos+dir] = field[x_pos][y_pos]
                        field[x_pos][y_pos] = 0
                        rand_gen = True
                    elif field[x_pos][y_pos+dir] == field[x_pos][y_pos] and field[x_pos][y_pos] !=0:
                        field[x_pos][y_pos+dir] = 2*field[x_pos][y_pos]
                        update_score(2*field[x_pos][y_pos])
                        field[x_pos][y_pos] = 0
                        rand_gen = True
            y_pos += dir
        x_pos += dir
    
    for i in counter:
        x_pos = 0
        for column in counter:
            y_pos = 0
            for row in counter:
                if field[x_pos][y_pos] != 0:
                    if limit_check(y_pos, dir):
                        if field[x_pos][y_pos+dir] == 0:
                            field[x_pos][y_pos+dir] = field[x_pos][y_pos]
                            field[x_pos][y_pos] = 0
                            rand_gen = True
                y_pos += dir
            x_pos += dir
    place_new()
```

```python

run = True


while run:
    # pygame.time.delay(100)
    draw_grid()
    draw_markers()
    draw_score()
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            run = False
            
        elif event.type == pygame.KEYUP:
            if event.key == pygame.K_d or event.key == pygame.K_RIGHT:
                move_horz(+1)

            if event.key == pygame.K_a or event.key == pygame.K_LEFT:    
                move_horz(-1)

            
            if event.key == pygame.K_w or event.key == pygame.K_UP:   
                move_vert(-1)

            
            if event.key == pygame.K_s or event.key == pygame.K_DOWN:   
                move_vert(+1)

            if event.key == pygame.K_r:
                empty_field()
            
    pygame.display.update()

pygame.quit()
```
