```python
line_width = 12

def draw_grid():
    grid = (187, 173, 160)
    for x in range(1,4):
        pg.draw.line(canvas, grid, (x0 + 0, y0 + x *100),
                         (x0 + 400, y0 + x * 100), line_width)
        pg.draw.line(canvas, grid, (x0 + x * 100, y0 + 0),
                         (x0 + x * 100, y0 + 400), line_width)
    pg.draw.line(canvas, grid, (x0, y0),
                         (x0 + 0, y0 + 400), 10)
    pg.draw.line(canvas, grid, (x0 + 400, y0 + 0),
                         (x0 + 400, y0 + 400), 10)
    pg.draw.line(canvas, grid, (x0 + 0, y0 + 0),
                         (x0 + 400, y0 + 0), 10)
    pg.draw.line(canvas, grid, (x0 + 0, y0 + 400),
                         (x0 + 400, y0 + 400), 10)
```
